# <p style="background-color:#e06666;font-family:newtimeroman;color:#ececec;font-size:120%;text-align:center;border-radius:40px 40px;">T-test</p>

**The indepentent T-test is a test used to check for a statistically significant difference in the means between 2 groups. It is also called the Student’s T-Test.**

**If you have less < 30 samples in each of the groups then you run the T-Test if you can assume the population follows a normal distribution.** 

# <p style="background-color:#e06666;font-family:newtimeroman;color:#ececec;font-size:120%;text-align:center;border-radius:40px 40px;">Types of T-test</p>

**We have 3 types of T-test.**

1. **Paired Sample T-Test. Compare the men of two samples taken from the same population but at different points in time.**

2. **Independent Sample T-Test. Compare the mean of two independent groups. You are comparing samples from two different populations.**

3. **One-Sample T-Test, we test if the mean of a single group is different from a known mean or hypothesized mean.**

# <p style="background-color:#e06666;font-family:newtimeroman;color:#ececec;font-size:120%;text-align:center;border-radius:40px 40px;">Example</p>

**In this example we will review with you Independent T-test.**
**I am going to compare height of Male and Female.**

**Also we need to discuss about hypotheses in this test.**
1. **Null hypothesis (H0) - the mean of sample 1 is equal to the mean of sample 2.**
2. **Alternative hypothesis (HA) - the mean of sample 1 is not equal to the mean of sample 2.**

```python
#Import libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import dataset
df = pd.read_csv("../input/height-of-male-and-female-by-country-2022/Height of Male and Female by Country 2022.csv")
```

```python
df.head()
```

```python
#cut only 'male Height' column for our data
#also cut only first 20 rows for our samle
maleHeight = df.iloc[:20, 2]
print(maleHeight)
```

```python
#cut only 'female Height' column for our data
#also cut only first 20 rows for our samle
femaleHeight = df.iloc[:20, 3]
print(femaleHeight)
```

```python
#let's plot this two smaples with their means.
maleMean = maleHeight.mean()
femaleMean = femaleHeight.mean()

plt.figure(figsize=(12,8))
plt.hist(maleHeight, bins=25, density=True, alpha=0.6, color='#07889b')
plt.hist(femaleHeight, bins=25, density=True, alpha=0.6, color='#e06666')

plt.axvline(maleMean, color='#00008B',lw=3, label=f'Male Mean: ${maleMean}')
plt.axvline(femaleMean, color='purple',lw=3, label=f'Female Mean: ${femaleMean}')

plt.xlabel('Heights')
plt.ylabel('Probability Density')
plt.title('Male vs Female', fontsize=16)
plt.legend()
```

**We can see from the graph that the means are different, but we need to make sure of this using our T-test.**

**The formula of Independent T-test is next:**




1. X1 - **mean of first sample**
2. X2 - **mean of second sample**
3. n1 - **size of the first sample**
4. n2 - **size of the second sample**
5. s1 - **standard deviation of the first sample.**
6. s2 - **standard deviation of the second sample.**


```python
#to calculate it we can use package 'stats' with his built-in 'ttest_ind' function.
from scipy import stats
tStat, pValue = stats.ttest_ind(maleHeight, femaleHeight, equal_var=True)
```

```python
print(pValue)
```

# <p style="background-color:#e06666;font-family:newtimeroman;color:#ececec;font-size:120%;text-align:center;border-radius:40px 40px;">Conclusion</p>

**Because the p-value (3.74) of our test  is greater than alpha = 0.05, we fail to reject the null hypothesis of the test. We do not have sufficient evidence to say that the Male mean height and Female mean height are different.**

if the p-value is less than .05, we reject the null hypothesis of the t-test and conclude that there is sufficient evidence to say that the two samples lead to different height mean between Male and Female.
